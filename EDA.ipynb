```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
import plotly.io as pio
pio.templates.default = "ggplot2"
```

```python
data = pd.read_csv("./data/local_currency.csv", encoding='CP949')
data.info()
```

```python
data['년월'] = [int(cnt[2:4]+cnt[5:7]) for cnt in data.기준년월]
```

```python
data['월별카드사용액/충전액'] = data['월별카드사용액(천원)']/data['월별카드충전액(천원)']
```

```python
data.isna().sum()
```

```python
data.value_counts('기준년월')
```

```python
fig = px.line(data, x="기준년월", y="월별카드발행수량(건)", color='시군명')
fig.show()
```

```python
fig = px.line(data, x="기준년월", y="월별카드충전액(천원)", color='시군명')
fig.show()
```

```python
fig = px.line(data, x="기준년월", y="월별카드사용액(천원)", color='시군명')
fig.show()
```
